### Load and preprocess data

- Load in the `rbf_dev_train.csv` and `rbf_dev_test.csv` train and test sets.
- For the train and test sets, separate the data variables (`X` and `Y`) from the class values (*class*). The test code below assumes the following names:
    - `y_train` and `y_test` for the class values in the train and test sets, respectively.
    - `rbf_dev_train` and `rbf_dev_test` for the train and test set data, respectively.
- Normalize each data feature "separately" between 0 and 1 (based on each feature's dynamic range).
    - Use the `min` and `max` values derived from the training set to normalize the test set. *This will ensure that identical feature values in the training and test sets get normalized to the same numeric value.*

```python
def normalization(train , test): 
    train_max_x = np.max(train[: , 0])
    train_min_x = np.min(train[: , 0])
    train_max_y = np.max(train[: , 1])
    train_min_y = np.min(train[: , 1])
    # print(train[: , :1])
    norm_train_x = (train[:, 0] - train_min_x) / (train_max_x - train_min_x)
    norm_train_y = (train[:, 1] - train_min_y) / (train_max_y - train_min_y)
    norm_train = np.column_stack((norm_train_x, norm_train_y))
    
    norm_test_x = (test[:, 0] - train_min_x) / (train_max_x - train_min_x)
    norm_test_y = (test[:, 1] - train_min_y) / (train_max_y - train_min_y)
    norm_test = np.column_stack((norm_test_x, norm_test_y))
    
    return norm_train, norm_test
```

```python
train = pd.read_csv('data/rbf_dev_train.csv')
test = pd.read_csv('data/rbf_dev_test.csv')

rbf_train = pd.DataFrame(train)
rbf_test = pd.DataFrame(test)

# print(rbf_train.shape)

rbf_dev_train_unorm = rbf_train.iloc[: , :2].to_numpy()
y_train = rbf_train.iloc[: , 2:].to_numpy().flatten()
rbf_dev_test_unorm = rbf_test.iloc[: , :2].to_numpy()
y_test = rbf_test.iloc[: , 2:].to_numpy().flatten()

# print(rbf_dev_train)
# print(y_train)
# print(rbf_dev_test)
# print(y_test)

rbf_dev_train , rbf_dev_test = normalization(rbf_dev_train_unorm , rbf_dev_test_unorm)
# print(rbf_dev_train[:5])
# print(rbf_dev_test[:5])
# y_train , y_test = normalization(y_train , y_test)
```

### Computing hidden and output layer activations

#### Testing hidden_act

```python
np.random.seed(1)

# N = 3 samples, M = 5 features
test_data = np.random.normal(size=(3, 5))
# 4 hidden units / prototypes, each has 5 features
test_centroids = np.random.normal(size=(4, 5))
# Each sample assigned to one of 4 hidden unit prototypes
test_sigmas = np.random.uniform(size=(4,))
test_wts = 2*np.random.uniform(size=(4+1, 3)) - 1

test_net = RBF_Net(4, 3)
test_net.prototypes = test_centroids
test_net.sigmas = test_sigmas
test_net.wts = test_wts
test_h_act = test_net.hidden_act(test_data)
print(f'Your hidden layer activation is\n{test_h_act}\n\nand should be')
print('[[0.      0.      0.00009 0.00033]\n [0.00013 0.      0.00004 0.00014]\n [0.      0.      0.      0.00001]]')
```

#### Testing output_act

```python
test_out_act = test_net.output_act(test_h_act)
print(f'Your output layer activation is\n{test_out_act}\n\nand should be')
print('[[-0.72136  0.61505 -0.20481]\n [-0.72151  0.61487 -0.20466]\n [-0.72144  0.61479 -0.20465]]')
```

```python
# Loading in the data and checking the shape
url = 'https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/master/diabetes.csv'

df_dbt = pd.read_csv(url)

dbt_x_train = df_dbt.iloc[:192 , :8].to_numpy()
dbt_y_train = df_dbt.iloc[:192 , -1].to_numpy()
dbt_x_test = df_dbt.iloc[192: , :8].to_numpy()
dbt_y_test = df_dbt.iloc[192: , -1].to_numpy()

print(f'Your training set shape is {dbt_x_train.shape}')
print(f'Your training classes shape is {dbt_y_train.shape}.')
print(f'Your test set shape is {dbt_x_test.shape}.')
print(f'Your test classes shape is {dbt_y_test.shape}.')
```

```python
# Creating a rbf object
dbt_num_hidden_units = 8
dbt_num_classes = len(np.unique(dbt_y_train))
dbt_net = RBF_Net(num_classes=dbt_num_classes, num_hidden_units=dbt_num_hidden_units)
```

```python
dbt_net.train(dbt_x_train, dbt_y_train)
print("Done")
```

```python
dbt_y_pred = dbt_net.predict(dbt_x_test)
```

```python
dbt_acc = dbt_net.accuracy(dbt_y_test , dbt_y_pred)
print("This is the accuracy:" , dbt_acc)
```

```python
# Creating a confusion matrix: 
dbt_confu_mat = dbt_net.confusion_matrix(dbt_y_test , dbt_y_pred, dbt_num_classes)
print(dbt_confu_mat)
```

```python
bar_x = np.unique(dbt_y_train)
height_pred = np.zeros(len(bar_x))
for i in range(len(bar_x)): 
    height_pred[i] = np.sum(dbt_y_pred == bar_x[i])

height_ori = np.zeros(len(bar_x))
for i in range(len(bar_x)): 
    height_ori[i] = np.sum(dbt_y_test == bar_x[i])

fig, axs = plt.subplots(1, 2 , figsize=(12,6))

# Plot original test values distribution
axs[0].bar(bar_x, height_ori, color=['skyblue', 'orange'])
axs[0].set_title('Original Test Values Distribution')
axs[0].set_xlabel('Class')
axs[0].set_ylabel('Count')

# Plot predicted values distribution
axs[1].bar(bar_x, height_pred, color=['skyblue', 'orange'])
axs[1].set_title('Predicted Values Distribution')
axs[1].set_xlabel('Class')
axs[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
